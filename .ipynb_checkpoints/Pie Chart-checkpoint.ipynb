```python
import matplotlib.pyplot as plt
```

```python
classes = ["python","R","Machine Learning","Artificial Intelligence","Data Science"]
class1_students = [45,15,35,25,30]
```

```python
plt.pie([1])
plt.show()
```

```python
plt.pie(class1_students)
plt.show()
```

```python
plt.pie(class1_students,labels=classes)
plt.show()
```

```python
plt.pie(class1_students,labels=classes,explode=[0,0,0,0,0])
plt.show()
```

```python
plt.pie(class1_students,labels=classes,explode=[0,0,0.2,0,0])
plt.show()
```

```python
plt.pie(class1_students,labels=classes,explode=[0.05,0,0.1,0,0])
plt.show()
```

```python
plt.pie(class1_students,labels=classes,explode=[0.05,0,0.1,0,0],startangle=90)
plt.show()
```

```python
plt.pie(class1_students,labels=classes,explode=[0.05,0,0.1,0,0],startangle=270)
plt.show()
```

```python
plt.pie(class1_students,labels=classes,explode=[0.05,0,0.1,0,0],startangle=270,colors=["c","r","g","b","m"])
plt.show()
```

```python
plt.pie(class1_students,labels=classes,explode=[0.05,0,0.1,0,0],startangle=270,colors=["c","r","g","b","m"],autopct="%0.2f%%")
plt.show()
```

```python
plt.pie(class1_students,labels=classes,explode=[0.05,0,0.1,0,0],startangle=270,
        colors=["c","r","g","b","m"],autopct="%0.2f%%",shadow=True)
plt.show()
```

```python
plt.pie(class1_students,labels=classes,explode=[0.05,0,0.1,0,0],startangle=270,
        colors=["c","r","g","b","m"],autopct="%0.2f%%",shadow=True,radius=1.5)
plt.show()
```

```python
plt.pie(class1_students,labels=classes,explode=[0.05,0,0.1,0,0],startangle=270,
        colors=["c","r","g","b","m"],autopct="%0.2f%%",shadow=True,radius=1.5,textprops={"fontsize":15})
plt.show()
```

```python
plt.pie(class1_students,labels=classes,explode=[0.05,0,0.1,0,0],startangle=270,
        colors=["c","r","g","b","m"],autopct="%0.2f%%",shadow=True,
        radius=1.5,textprops={"fontsize":15},center=(0,0,))
plt.show()
```
