```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
ml_students_age = np.random.randint(18,45,(100))
py_students_age = np.random.randint(15,40,(100))
```

```python
print(ml_students_age)
print(py_students_age)
```

```python
plt.hist(ml_students_age)
```

```python
plt.hist(ml_students_age)
plt.title("ML Students age histogram")
plt.xlabel("Students age category")
plt.ylabel("No. Students age")
```

```python
plt.hist(ml_students_age)
plt.title("ML Students age histogram",fontsize=15)
plt.xlabel("Students age category",fontsize=15)
plt.ylabel("No. Students age",fontsize=15)
```

```python
bins=[15,20,25,30,35,40,45]
plt.hist(ml_students_age,bins)
plt.title("ML Students age histogram",fontsize=15)
plt.xlabel("Students age category",fontsize=15)
plt.ylabel("No. Students age",fontsize=15)
```

```python
bins=[15,20,25,30,35,40,45]
plt.hist(ml_students_age,bins,rwidth=0.8)
plt.title("ML Students age histogram",fontsize=15)
plt.xlabel("Students age category",fontsize=15)
plt.ylabel("No. Students age",fontsize=15)
```

```python
bins=[15,20,25,30,35,40,45]
plt.hist(ml_students_age,bins,rwidth=0.8,color="r")
plt.title("ML Students age histogram",fontsize=15)
plt.xlabel("Students age category",fontsize=15)
plt.ylabel("No. Students age",fontsize=15)
```

```python
bins=[15,20,25,30,35,40,45]
plt.hist(ml_students_age,bins,rwidth=0.8,orientation="horizontal")
plt.title("ML Students age histogram",fontsize=15)
plt.xlabel("Students age category",fontsize=15)
plt.ylabel("No. Students age",fontsize=15)
```

```python
bins=[15,20,25,30,35,40,45]
plt.hist(ml_students_age,bins,rwidth=0.8,color="r",label="Ml_Student_age")
plt.title("ML Students age histogram",fontsize=15)
plt.xlabel("Students age category",fontsize=15)
plt.ylabel("No. Students age",fontsize=15)
plt.legend()
```

```python
bins=[15,20,25,30,35,40,45]
plt.hist(ml_students_age,bins,rwidth=0.8,color="r",histtype="step",label="Ml_Student_age")
plt.title("ML Students age histogram",fontsize=15)
plt.xlabel("Students age category",fontsize=15)
plt.ylabel("No. Students age",fontsize=15)
plt.legend()
```

```python
bins=[15,20,25,30,35,40,45]
plt.hist(ml_students_age,bins,rwidth=0.8,color="r",histtype="stepfilled",label="Ml_Student_age")
plt.title("ML Students age histogram",fontsize=15)
plt.xlabel("Students age category",fontsize=15)
plt.ylabel("No. Students age",fontsize=15)
plt.legend()
```

```python
bins=[15,20,25,30,35,40,45]
plt.hist(ml_students_age,bins,rwidth=0.8,color="r",histtype="bar",label="Ml_Student_age")
plt.title("ML Students age histogram",fontsize=15)
plt.xlabel("Students age category",fontsize=15)
plt.ylabel("No. Students age",fontsize=15)
plt.legend()
```

```python
bins=[15,20,25,30,35,40,45]
plt.hist(ml_students_age,bins,rwidth=0.8,color="r",histtype="barstacked",label="Ml_Student_age")
plt.title("ML Students age histogram",fontsize=15)
plt.xlabel("Students age category",fontsize=15)
plt.ylabel("No. Students age",fontsize=15)
plt.legend()
```

```python
bins=[15,20,25,30,35,40,45]
plt.hist(ml_students_age,bins,rwidth=0.8,color="r",histtype="stepfilled",label="Ml_Student_age",cumulative=True)
plt.title("ML Students age histogram",fontsize=15)
plt.xlabel("Students age category",fontsize=15)
plt.ylabel("No. Students age",fontsize=15)
plt.legend()
```

```python
bins=[15,20,25,30,35,40,45]
plt.hist(ml_students_age,bins,rwidth=0.8,color="r",histtype="bar",label="Ml_Student_age")
plt.hist(py_students_age,bins,rwidth=0.8,color="g",histtype="bar",label="py_Student_age")
plt.title("ML Students age histogram",fontsize=15)
plt.xlabel("Students age category",fontsize=15)
plt.ylabel("No. Students age",fontsize=15)
plt.legend()
```

```python
bins=[15,20,25,30,35,40,45]
plt.hist(ml_students_age,bins,orientation="vertical",rwidth=0.8,color="r",histtype="bar",label="Ml_Student_age")
plt.hist(py_students_age,bins,orientation="horizontal",rwidth=0.8,color="g",histtype="bar",label="py_Student_age")
plt.title("ML Students age histogram",fontsize=15)
plt.xlabel("Students age category",fontsize=15)
plt.ylabel("No. Students age",fontsize=15)
plt.legend()
```

```python
bins=[15,20,25,30,35,40,45]
plt.hist([ml_students_age,py_students_age],bins,rwidth=0.8,color=["r","g"],histtype="bar",label=["Ml_Student_age","py_Student_age"])
plt.title("ML Students age histogram",fontsize=15)
plt.xlabel("Students age category",fontsize=15)
plt.ylabel("No. Students age",fontsize=15)
plt.legend()
```

```python
import matplotlib.pyplot as plt
from matplotlib.pyplot import style
bins=[15,20,25,30,35,40,45]
style.use("ggplot")
plt.hist([ml_students_age,py_students_age],bins,rwidth=0.8,color=["r","g"],histtype="bar",label=["Ml_Student_age","py_Student_age"])
plt.title("ML Students age histogram",fontsize=15)
plt.xlabel("Students age category",fontsize=15)
plt.ylabel("No. Students age",fontsize=15)
plt.legend()
```

```python
import matplotlib.pyplot as plt
from matplotlib.pyplot import style
bins=[15,20,25,30,35,40,45]
style.use("ggplot")
plt.figure(figsize=(16,9))
plt.hist([ml_students_age,py_students_age],bins,rwidth=0.8,color=["r","g"],histtype="bar",label=["Ml_Student_age","py_Student_age"])
plt.title("ML Students age histogram",fontsize=15)
plt.xlabel("Students age category",fontsize=15)
plt.ylabel("No. Students age",fontsize=15)
plt.legend()
```

```python
import matplotlib.pyplot as plt
from matplotlib.pyplot import style
bins=[15,20,25,30,35,40,45]
style.use("ggplot")
plt.figure(figsize=(16,9))
plt.hist([ml_students_age,py_students_age],bins,rwidth=0.8,color=["r","g"],histtype="bar",label=["Ml_Student_age","py_Student_age"])
plt.title("ML Students age histogram",fontsize=15)
plt.xlabel("Students age category",fontsize=15)
plt.ylabel("No. Students age",fontsize=15)
plt.legend()
plt.grid(color="w",linewidth=2,linestyle="-.")
```

```python
import matplotlib.pyplot as plt
from matplotlib.pyplot import style
bins=[15,20,25,30,35,40,45]
style.use("ggplot")
plt.figure(figsize=(16,9))
plt.hist([ml_students_age,py_students_age],bins,rwidth=0.8,color=["r","g"],histtype="bar",label=["Ml_Student_age","py_Student_age"])
plt.title("ML Students age histogram",fontsize=15)
plt.xlabel("Students age category",fontsize=15)
plt.ylabel("No. Students age",fontsize=15)
plt.legend()
plt.grid(color="w",linewidth=2,linestyle="--")
```
