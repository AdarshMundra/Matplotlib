```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib import style
```

```python
classes = ["python", "R" , "AI" , "ML" , "DS"]
class1_students = [30,10,20,25,10]
class2_students=[40,5,20,20,10]
class3_students= [35,5,30,15,15]
```

```python
plt.bar(classes,class1_students)
```

```python
plt.barh(classes,class1_students)
```

```python
plt.bar(classes,class1_students,color="r")
```

```python
plt.bar(classes,class1_students,color="r",align="edge")
```

```python
plt.bar(classes,class1_students,color="r",align="edge",edgecolor="g")
```

```python
plt.bar(classes,class1_students,color="r",align="edge",edgecolor="g",linestyle="--")
```

```python
plt.bar(classes,class1_students,color="r",align="edge",edgecolor="g",linestyle="--",linewidth=4)
```

```python
plt.bar(classes,class1_students,color="r",align="edge",edgecolor="g",linestyle="--",linewidth=4,width=0.4,alpha=0.7,label="Class 1 Students")
```

```python
plt.bar(classes,class1_students,color="r",align="edge",edgecolor="g",linestyle="--",
        linewidth=2,width=0.6,alpha=0.7,label="Class 1 Students")
```

```python
style.use("ggplot")
plt.bar(classes,class1_students,color="r",align="edge",edgecolor="g",linestyle="--",
        linewidth=2,width=0.4,alpha=0.7,label="Class 1 Students")
```

```python
style.use("ggplot")
plt.figure(figsize=(16,9))
plt.bar(classes,class1_students,color="r",align="edge",edgecolor="g",linestyle="--",
        linewidth=2,width=0.4,alpha=0.7,label="Class 1 Students")
```

```python
plt.figure(figsize=(16,9))
plt.bar(classes,class1_students,color="r",align="edge",edgecolor="g",linestyle="--",
        linewidth=2,width=0.4,alpha=0.7,label="Class 1 Students")
plt.title("Bar Chart of IAIP Class",fontsize=1)
plt.xlabel("Classes")
plt.ylabel("No. of Students")
```

```python
plt.figure(figsize=(16,9))
plt.bar(classes,class1_students,color="r",align="edge",edgecolor="g",linestyle="--",
        linewidth=2,width=0.4,alpha=0.7,label="Class 1 Students")

plt.bar(classes,class2_students,color="g",align="edge",edgecolor="r",linestyle="--",
        linewidth=2,width=0.4,label="Class 2 Students")

plt.title("Bar Chart of IAIP Class",fontsize=20)
plt.xlabel("Classes",fontsize=15)
plt.ylabel("No. of Students",fontsize=15)
plt.legend()
```

```python
plt.figure(figsize=(16,9))
classes_index = np.arange(len(classes))
width=0.2
plt.bar(classes_index,class1_students,width,color="r",align="edge",edgecolor="g",linestyle="--",
        linewidth=2,alpha=0.7,label="Class 1 Students")

plt.bar(classes_index,class2_students,width,color="g",align="edge",edgecolor="r",linestyle="--",
        linewidth=2,label="Class 2 Students")

plt.title("Bar Chart of IAIP Class",fontsize=20)
plt.xlabel("Classes",fontsize=15)
plt.ylabel("No. of Students",fontsize=15)
plt.legend()
```

```python
plt.figure(figsize=(16,9))
classes_index = np.arange(len(classes))
width=0.2
plt.bar(classes_index,class1_students,width,color="r",align="edge",edgecolor="g",linestyle="--",
        linewidth=2,alpha=0.7,label="Class 1 Students")

plt.bar(classes_index + width,class2_students,width,color="g",align="edge",edgecolor="r",linestyle="--",
        linewidth=2,label="Class 2 Students")

plt.title("Bar Chart of IAIP Class",fontsize=20)
plt.xlabel("Classes",fontsize=15)
plt.ylabel("No. of Students",fontsize=15)
plt.legend()
```

```python
plt.figure(figsize=(16,9))
classes_index = np.arange(len(classes))
width=0.2
plt.bar(classes_index,class1_students,width,color="r",align="edge",edgecolor="g",linestyle="--",
        linewidth=2,alpha=0.7,label="Class 1 Students")

plt.bar(classes_index + width,class2_students,width,color="g",align="edge",edgecolor="r",linestyle="--",
        linewidth=2,label="Class 2 Students")

plt.bar(classes_index + width + width,class3_students,width,color="b",align="edge",edgecolor="y",linestyle="--",
        linewidth=2,label="Class 3 Students")

plt.title("Bar Chart of IAIP Class",fontsize=20)
plt.xlabel("Classes",fontsize=15)
plt.ylabel("No. of Students",fontsize=15)
plt.legend()
```

```python
plt.figure(figsize=(16,9))
classes_index = np.arange(len(classes))
width=0.2
plt.bar(classes_index,class1_students,width,color="r",align="edge",edgecolor="g",linestyle="--",
        linewidth=2,alpha=0.7,label="Class 1 Students")

plt.bar(classes_index + width,class2_students,width,color="g",align="edge",edgecolor="r",linestyle="--",
        linewidth=2,label="Class 2 Students")

plt.bar(classes_index + width + width,class3_students,width,color="b",align="edge",edgecolor="y",linestyle="--",
        linewidth=2,label="Class 3 Students")
plt.xticks(classes_index + width,classes)
plt.title("Bar Chart of IAIP Class",fontsize=20)
plt.xlabel("Classes",fontsize=15)
plt.ylabel("No. of Students",fontsize=15)
plt.legend()
```

```python
plt.figure(figsize=(16,9))
classes_index = np.arange(len(classes))
width=0.2
plt.bar(classes_index,class1_students,width,color="r",align="edge",edgecolor="g",linestyle="--",
        linewidth=2,alpha=0.7,label="Class 1 Students")

plt.bar(classes_index + width,class2_students,width,color="g",align="edge",edgecolor="r",linestyle="--",
        linewidth=2,label="Class 2 Students")

plt.bar(classes_index + width + width,class3_students,width,color="b",align="edge",edgecolor="y",linestyle="--",
        linewidth=2,label="Class 3 Students")
plt.xticks(classes_index + width,classes,rotation=20)
plt.title("Bar Chart of IAIP Class",fontsize=20)
plt.xlabel("Classes",fontsize=15)
plt.ylabel("No. of Students",fontsize=15)
plt.legend()
```
