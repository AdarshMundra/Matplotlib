```python
plt.figure(figsize=(16,9))
plt.scatter(x,y,c="r",marker="*",s=125)
plt.title("Google Play Store Data",fontsize=15)
plt.xlabel("Rating",fontsize=12)
plt.ylabel("Review",fontsize=12)
```

```python
plt.figure(figsize=(16,9))
plt.scatter(x,y,c="r",marker="*",s=125,alpha=0.6)
plt.title("Google Play Store Data",fontsize=15)
plt.xlabel("Rating",fontsize=12)
plt.ylabel("Review",fontsize=12)
```

```python
plt.figure(figsize=(16,9))
plt.scatter(x,y,c="r",marker="*",s=125,linewidth=6)
plt.title("Google Play Store Data",fontsize=15)
plt.xlabel("Rating",fontsize=12)
plt.ylabel("Review",fontsize=12)
```

```python
plt.figure(figsize=(16,9))
plt.scatter(x,y,c="r",marker="*",s=125,linewidth=6,edgecolor="b")
plt.title("Google Play Store Data",fontsize=15)
plt.xlabel("Rating",fontsize=12)
plt.ylabel("Review",fontsize=12)
```

```python
plt.figure(figsize=(16,9))
plt.scatter(x,y,c="r",marker="*",s=125,linewidth=6,edgecolor="b")
z=df["Installs"]
plt.scatter(x,z,c="y",marker="o",s=120,linewidth=4,edgecolor="c")
plt.title("Google Play Store Data",fontsize=25)
plt.xlabel("Rating",fontsize=20)
plt.ylabel("Review & Installs",fontsize=20)
```

```python
plt.figure(figsize=(16,9))
plt.scatter(x,y,c="r",marker="*",s=125,linewidth=6,edgecolor="b",label="Reviews")
z=df["Installs"]
plt.scatter(x,z,c="y",marker="o",s=120,linewidth=4,edgecolor="c",label="Install")
plt.title("Google Play Store Data",fontsize=25)
plt.xlabel("Rating",fontsize=20)
plt.ylabel("Review & Installs",fontsize=20)
plt.legend()
```
